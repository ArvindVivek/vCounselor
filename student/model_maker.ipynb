```python
import pandas as pd
import numpy as np

data_math = pd.read_csv('student-mat.csv')
data_port = pd.read_csv('student-por.csv')
```

```python
data_math.head()
```

```python
data_port.head()
```

```python
data_math = data_math['school;sex;age;address;famsize;Pstatus;Medu;Fedu;Mjob;Fjob;reason;guardian;traveltime;studytime;failures;schoolsup;famsup;paid;activities;nursery;higher;internet;romantic;famrel;freetime;goout;Dalc;Walc;health;absences;G1;G2;G3'].apply(lambda x: pd.Series(x.split(';')))
data_port = data_port['school;sex;age;address;famsize;Pstatus;Medu;Fedu;Mjob;Fjob;reason;guardian;traveltime;studytime;failures;schoolsup;famsup;paid;activities;nursery;higher;internet;romantic;famrel;freetime;goout;Dalc;Walc;health;absences;G1;G2;G3'].apply(lambda x: pd.Series(x.split(';')))
```

```python
cols = ['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc','health', 'absences', 'G1', 'G2', 'G3' ]
data_math.columns = cols
data_port.columns = cols
```

```python
data_math.head()
```

```python
data_port.head()
```

```python
scores = ['G1', 'G2', 'G3']
for i in range(len(data_math)):
    for j in scores:
        data_math[j][i] = data_math[j][i].replace('"', "")
        data_port[j][i] = data_port[j][i].replace('"', "")
        data_math[j][i] = (float(data_math[j][i]) + float(data_port[j][i]))//2
data = data_math
```

```python
data.head()
```

```python
print(data.columns)
dropped_cols = ['school','nursery', 'romantic', 'Dalc', 'Walc', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason']
data.drop(dropped_cols, inplace = True, axis = 1)
```

```python
missing_cols = [i for i in data.columns if data[i].isnull().any()]
missing_cols
```

```python
cat_cols = [i for i in data.columns if data[i].dtypes == 'O']
cat_cols.remove('G1')
cat_cols.remove('G2')
cat_cols.remove('G3')

cat_cols
```

```python
def label_encode(col):
    u = list(data[col].unique())
    n = []
    for i in range(len(data)):
        n.append(u.index(data[col][i]))
    return n

for i in cat_cols:
    data[i] = label_encode(i)
        
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

data.drop(['G1', 'G2'], inplace = True, axis = 1)
features = list(data.columns)
features.remove('G3')

X = data[features]
y = data['G3']

X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

```python
model = RandomForestRegressor(n_estimators = 100, random_state = 1)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_valid)
mae = mean_absolute_error(y_valid, pred)
print("MAE:", mae)
```
